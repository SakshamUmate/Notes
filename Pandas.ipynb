# Pandas


Data Input/Output:

read_csv(): Read a CSV file into a DataFrame.
read_excel(): Read an Excel file into a DataFrame.
to_csv(): Write a DataFrame to a CSV file.
to_excel(): Write a DataFrame to an Excel file.

Data Exploration and Display:

head(): Display the first few rows of a DataFrame.
tail(): Display the last few rows of a DataFrame.
info(): Display a concise summary of a DataFrame.
describe(): Generate descriptive statistics of a DataFrame.
shape: Return the dimensions (rows, columns) of a DataFrame.m

Data Selection and Indexing:

loc[]: Access a group of rows and columns by labels.
iloc[]: Access a group of rows and columns by integer position.
at[]: Access a single value for a row/column label pair.
iat[]: Access a single value for a row/column pair by integer position.

Data Manipulation:

drop(): Remove specified labels from rows or columns.
groupby(): Group a DataFrame by one or more columns.
merge(): Merge two DataFrames on a specified column or index.
concat(): Concatenate two or more DataFrames along rows or columns.
pivot_table(): Create a spreadsheet-style pivot table as a DataFrame.

Data Cleaning:

fillna(): Fill missing values in a DataFrame.
dropna(): Remove missing values from a DataFrame.
replace(): Replace specified values with other values.

Statistical Functions:

mean(): Calculate the mean of a DataFrame or Series.
sum(): Calculate the sum of values in a DataFrame or Series.
min(), max(): Find the minimum or maximum value in a DataFrame or Series.
