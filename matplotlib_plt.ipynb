```python
x=np.linspace(0,10,50)
y=np.sin(x)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--g',)
plt.title("this is sin graph")
plt.xlabel("this is my value 0-10")
plt.ylabel("this is sin(X)")
plt.grid()
plt.show()
```
