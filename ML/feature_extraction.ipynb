# Feature Scaling: $ Standardization $ 

1. means apllying $ z= \frac {\bar X - \mu}{\sigma} $ to each feature, where $\bar X$ is the feature value, $\mu$ is the mean of the feature and $\sigma$ is the standard deviation of the feature.

```python
from sklearn.preprocessing import StandardScaler
```

```python
import seaborn as sns
import pandas as pd 
```

```python
df=sns.load_dataset("tips")
```

```python
df.head(3)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[["total_bill"]])
```

```python
df1=pd.DataFrame(scaler.transform(df[["total_bill"]]))
```

```python
sns.histplot(df1)
```

```python
sns.histplot(df["total_bill"])  
```

# $ (Min-Max Scaling) $  $ Normalization $
### [ Min-Max Scaling ] is another method of feature scaling. It scales the feature values between 0 and 1.

2. means apllying $ X_{norm} = \frac {X - X_{min}}{X_{max} - X_{min}} $ to each feature, where $X_{min}$ is the minimum value of the feature and $X_{max}$ is the maximum value of the feature.

```python
from sklearn.preprocessing import  MinMaxScaler

minmax_scaler = MinMaxScaler()
```

```python
df.head(2)
```

```python
minmax_scaler.fit(df[['distance','total']])
```

```python
minmax_scaler.transform(df[['distance','total']])
```

this atributte can do fit and transform in the same time

```python
minmax_scaler.fit_transform(df[['distance','total']])
```
