```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
import pandas as pd

df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
```

```python
final=pd.concat([df1,df2],axis=1)
```

```python
final['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(final['f1'],final['f2'],c=final['target'])
```

```python
from imblearn.over_sampling import  SMOTE
```

```python
over_sample=SMOTE()
x,y=over_sample.fit_resample(final[['f1','f2']],final['target'])
```

```python
def Count(X,Y):
    if X==0:
        return 1
    if Y==0:
        return 1
    else:
        return Count(X-1,Y) + Count(X-1,Y-1)
```
