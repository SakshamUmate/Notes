```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,50)
y=np.sin(x)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sin(x)')
plt.grid()
plt.plot(x,y,'+',color='blue')
```

```python
import matplotlib.pyplot as plt
import numpy as np

data_heatmap=np.random.rand(10,10)
plt.imshow(data_heatmap, cmap='viridis', interpolation='nearest')
plt.title('Heatmap')
plt.colorbar()
plt.show()
```

```python
data_area = np.vstack([np.arange(10), np.random.rand(10)]).T
plt.fill_between(data_area[:, 0], data_area[:, 1], color='skyblue', alpha=0.5)
plt.title('Area Chart')
plt.show()
```

```python
x=np.random.randint(0, 10, 20)
y=np.random.randint(0, 10, 20)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.grid()
plt.scatter(x,y,marker='^',alpha=0.5,color='red')
```

```python
x=np.random.randint(0, 10, 50)
plt.hist(x, bins=10, color='b', alpha=0.6)
plt.xlabel('Numbers in x Array')
plt.title('Histogram')  
plt.show()
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.grid()
sizes=np.random.randint(0,100,50)
plt.scatter(x, y, marker='o', c='blue', alpha=0.5, sizes=sizes)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np



The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid, specifying the number of rows and columns, and then selecting a specific subplot to draw on. This is useful when you want to compare different datasets or visualize multiple aspects of your data side by side.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the four lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.grid()
plt.plot(x,y1)
plt.title('ploting y1')
plt.xlabel('x')
plt.ylabel('y1')

plt.subplot(2, 2, 2)
plt.grid()
plt.plot(x,y2,color='red')
plt.title('ploting y2')
plt.xlabel('x')
plt.ylabel('y2')

plt.subplot(2, 2, 3)
plt.grid()
plt.plot(x,y3,color='green')
plt.title('ploting y3')

plt.subplot(2, 2, 4)
plt.grid()
plt.plot(x,y4,color='orange')
plt.title('ploting y4') 

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.title('Bar Chart')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.grid()
plt.bar(company, profit, color='skyblue')
```

```python
plt.barh(company, profit, color='aqua')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Horizontal  Bar Chart')
```
