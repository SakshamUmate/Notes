```python
def bin_search(L,x):
    l=0
    h=len(L)-1
    while l<=h:
        mid=(l+h)//2
        if L[mid]==x:
            return True
        if L[mid]<x:
            l=mid+1
        else:
            h=mid-1
    return False

```
