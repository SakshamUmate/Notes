# $ Linear $ $ interpolation $


```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x=np.linspace(0,6,5)
y=np.linspace(0,6,5)
```

```python
plt.scatter(x,y)
```

```python
x_new=np.linspace(0,6,10)
y_inter=np.interp(x_new,x,y)
```

```python
plt.scatter(x_new,y_inter)
```

# $Cubic $ $interpolaton$ 

```python

x=[1,2,3,4,5]
y=[1,8,27,64,125]
plt.scatter(x,y)
```

```python
from scipy.interpolate import  interp1d
```

```python
fun=interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1,5,10)
y_inter = fun(x_new)
```

```python
plt.scatter(x_new,y_inter,c='black',marker='>')
```

# $ Polynomial $ $ interpolation $

```python
np.random.seed(32)
x = np.linspace(1,15,5)
y = [i**2 for i in x]

```

```python
plt.scatter(x,y)
```

```python
p=np.polyfit(x,y,2)
x_new = np.linspace(1,15,10)
y_inter =np.polyval(p,x_new)
```

```python
plt.scatter(x_new,y_inter)
```
